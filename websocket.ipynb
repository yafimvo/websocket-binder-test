```python
import threading
import asyncio
import websockets

async def websocket_handler(websocket, path):
    await websocket.send("from socket")

def start_websocket_server():
    asyncio.set_event_loop(asyncio.new_event_loop())
    start_server = websockets.serve(websocket_handler, "0.0.0.0", 1234)
    
    asyncio.get_event_loop().run_until_complete(start_server)
    port = start_server.ws_server.sockets[0].getsockname()[1]
    print(f"{port}")      
    asyncio.get_event_loop().run_forever()

# Create a new thread and start the WebSocket server in that thread
websocket_thread = threading.Thread(target=start_websocket_server)
websocket_thread.start()
```
